```python
#Create a copy of the original dataset to work on
data=data2.copy()
data
```

```python
#pd.set_option("display.max_rows", None)
data
```

```python
#convert Temp C which is object to Float(numerical)
data['Temp C']=pd.to_numeric(data['Temp C'],errors='coerce')
data.dtypes
```

```python
#Convert Weather from object to category and wind from flot to int64
data['Weather']=data['Weather'].astype('category')
data['Wind']=data['Wind'].astype('int64')
data.dtypes
```

```python
#drop duplicates
data_cleaned1=data.drop_duplicates()
```

```python
#new shape
data_cleaned1.shape
```

```python
#Drop tempC col as it is duplicate
data_cleaned2=data_cleaned1.drop('Temp C',axis=1)
data_cleaned2
```

```python
#Rename Solar R to Solar
data_cleaned3 = data_cleaned2.rename({'Solar.R': 'Solar'}, axis=1)
data_cleaned3
#Create Histogram on Ozone col
data_cleaned3['Ozone'].hist()
```

```python
#Create Heatmap to identify null values
import seaborn as sns
cols = data_cleaned3.columns
colours = ['#000099', '#ffff00'] # specify the colours - yellow is missing. blue is not missing.
sns.heatmap(data_cleaned3[cols].isnull(),
            cmap=sns.color_palette(colours))
```

```python
#Null Value summary with cols and head fn
data_cleaned3[data_cleaned3.isnull().any(axis=1)].head()
```

```python
#Null values summary
data_cleaned3.isnull().sum()
```

```python
#replace null values with mean vlaues
mean = data_cleaned3['Ozone'].mean()
print(mean)
```

```python
data_cleaned3['Ozone'] = data_cleaned3['Ozone'].fillna(mean)
```

```python
data_cleaned3['Ozone'].isnull().sum()
```

```python
data_cleaned3['Ozone']
```

```python
obj_columns=data_cleaned3[['Weather']]
```

```python
obj_columns.value_counts()
```

```python
#Replace null values in weather col with mode
obj_columns=obj_columns.fillna(obj_columns.mode().iloc[0])
```

```python
#check if there are any null values after mode imputation
obj_columns.isnull().sum()
```

```python
#Replace null values in Solar with mean
data_cleaned3['Solar'] = data_cleaned3['Solar'].fillna(mean)
```

```python
#Drop weather col to learn concat
data_cleaned3=data_cleaned3.drop('Weather',axis=1)
data_cleaned3
```

```python
data_cleaned3
```

```python
#Concat weather col
data_cleaned3 = pd.concat([data_cleaned3,obj_columns], axis=1)
data_cleaned3 .isnull().sum()
```

```python
# Seaborn visualization library
import seaborn as sns
# Create the default pairplot
sns.pairplot(data_cleaned3.iloc[:,0:7])
```

```python
#Drop year col as it is same .variance is zero
data_cleaned3=data_cleaned3.drop('Year',axis=1)
#Check Correlation
data_cleaned3.corr()
```

```python
#Creating dummy variable for Weather column since the weather col is categorical
data_cleaned4=pd.get_dummies(data,columns=['Weather'])
data_cleaned4
```

```python
#Already dropped NA values
data_cleaned4=data_cleaned4.dropna()
data_cleaned4
```

```python
#Month col has 1 non numercial value(may)replace that with 5
data_cleaned4 ['Month'] = data_cleaned4 ['Month'].replace(['May'],'5')
data_cleaned4
```

```python
#Normalization of the data
from numpy import set_printoptions
from sklearn.preprocessing import MinMaxScaler
array = data_cleaned4.values

scaler = MinMaxScaler(feature_range=(0,1))
rescaledX = scaler.fit_transform(array[:,0:4])

#transformed data precicion =2 for decimal pt 2
set_printoptions(precision=2)
print(rescaledX[0:5,:])
```

```python
#Standardization convert all to Z score
from sklearn.preprocessing import StandardScaler
array = data_cleaned4.values
scaler = StandardScaler().fit(array)
rescaledX = scaler.transform(array)

# summarize transformed data
set_printoptions(precision=2)
print(rescaledX[0:5,0:4])
```
