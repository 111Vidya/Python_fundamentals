```python
#User Based Collaborative Filtering
import pandas as pd
import numpy as np
```

```python
#Sort based on used ID and reset index
movies_df = pd.read_csv('Movie (1).csv')
movies_df = movies_df.sort_values(by ='userId').reset_index(drop = True)
movies_df
```

```python
movies_df[0:5]
```

```python
#number of unique users in the dataset
len(movies_df.userId.unique())
```

```python
#number of unique movies in the dataset
len(movies_df.movie.unique())
```

```python
#User based collaborative filterngs.User item interaction matrix. pivot fn col shud be movie names and rows user ID
user_movies_df = movies_df.pivot(index='userId',
                                 columns='movie',
                                 values='rating').reset_index()
```

```python
user_movies_df
```

```python
#replace index with user ID unque values. we dont need index in a series , we need the user ID the actual user number
user_movies_df.index = movies_df.userId.unique()
```

```python
user_movies_df
```

```python
#Impute those NaNs with 0 values
user_movies_df.fillna(0, inplace=True)
```

```python
user_movies_df
```

```python
#Calculating Cosine Similarity between Users-distances library. distances b/w obs
from sklearn.metrics import pairwise_distances
#from scipy.spatial.distance import cosine, correlation
```

```python
#Pairwaise distance fn metric cosine - cosine similarity. why is it 1-?
#Pairwise will calc dissimilarty, 1- will give similarity .will give us similarity matrix
user_sim = 1 - pairwise_distances( user_movies_df.values,metric='cosine')
```

```python
#All diagonal values are 1 correlation with itself.Similarity Matrix
user_sim
```

```python
#Store the results in a dataframe
user_sim_df = pd.DataFrame(user_sim)
```

```python
#Set the index and column names to user ids
user_sim_df.index = movies_df.userId.unique()
user_sim_df.columns = movies_df.userId.unique()
```

```python
user_sim_df.iloc[0:5, 0:5]
```

```python
#Diagonal information is not important. replace them as 0
np.fill_diagonal(user_sim, 0)
user_sim_df.iloc[0:5, 0:5]
```

```python
#from the remaining matrix, Most Similar Users.idxmax. Cosine vlaue is high b/w UserID 1 and user ID 5. 1& 5 are similar users
user_sim_df.idxmax(axis=1)[0:5]
```

```python
#Item Based Collaborative Filtering. collect ratings for Sabrina
#Fetching ratings for Toy Story
sabrina_user_rating = user_movies_df['Sabrina (1995)']
```

```python
#All the ratings for Sabrina
sabrina_user_rating
```

```python
#Finding the correlation of Sabrina with different movies
similar_to_sabrina = user_movies_df.corrwith(sabrina_user_rating)
```

```python
#Look for high positive corelation for movies with Sabrina
similar_to_sabrina.head(10)
```

```python
#Look for high positive corelation for movies with Sabrina
corr_sabrina = pd.DataFrame(similar_to_sabrina, columns=['Correlation'])
corr_sabrina.dropna(inplace=True)
corr_sabrina.head()
```

```python
# Recommend these movies with high correaltion to users who watched Sabrina. sort high +ve correlation.
corr_sabrina[corr_sabrina['Correlation'] > 0].sort_values(by='Correlation', ascending=False).head(5)
```
