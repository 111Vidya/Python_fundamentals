```python
beml_df=pd.read_csv("BEML.csv")
beml_df[0:5]
```

```python
glaxo_df=pd.read_csv('GLAXO.csv')
glaxo_df[0:5]
```

```python
beml_df=beml_df[['Date','Close']]
glaxo_df=glaxo_df[['Date','Close']]
```

```python
beml_df
```

```python
glaxo_df
```

```python
#The data frames have a date column, so we can create aDatetimeIndexfrom the Date.
# Rows are sorted by time in ascending Order
beml_df=beml_df.set_index(pd.DatetimeIndex(beml_df['Date']))
glaxo_df=glaxo_df.set_index(pd.DatetimeIndex(beml_df['Date']))
```

```python
beml_df
```

```python
glaxo_df
```

```python
import matplotlib.pyplot as plt
import seaborn as sn
plt.plot(glaxo_df.Close)
plt.xlabel('Time')
plt.ylabel('Close Price')
plt.show()
```

```python
plt.plot(beml_df.Close)
plt.xlabel('Time')
plt.ylabel('Close Price')
plt.show()
```

```python
glaxo_df['gain']=glaxo_df.Close.pct_change(periods=1)
beml_df['gain']=beml_df.Close.pct_change(periods=1)
beml_df
glaxo_df
```

```python
glaxo_df=glaxo_df.dropna()
beml_df=beml_df.dropna()
glaxo_df
```

```python
glaxo_df=glaxo_df.dropna()
beml_df=beml_df.dropna()
glaxo_df.isnull().sum()
```

```python
plt.figure(figsize=(10,5))
plt.plot(glaxo_df.gain)
plt.xlabel('Time')
plt.ylabel('gain')
```

```python
import warnings
warnings.filterwarnings('ignore')
sn.distplot(glaxo_df.gain,label="Glaxo")
plt.legend()
plt.show()
```

```python
sn.distplot(beml_df.gain,label="BEML",color='r')
plt.legend()
plt.show()
```

```python
x=pd.DataFrame(columns=["beml_df.gain","glaxo_df.gain"])
x["beml_df.gain"]=pd.Series(beml_df.gain)
x["glaxo_df.gain"]=pd.Series(glaxo_df.gain)
sn.distplot(x['glaxo_df.gain'],label='Glaxo')
sn.distplot(x["beml_df.gain"],label="BEML")
plt.xlabel("gain")
plt.ylabel("density")
plt.legend()
```
