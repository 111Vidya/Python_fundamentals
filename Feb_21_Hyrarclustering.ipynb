```python
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
Univ = pd.read_csv("Universities.csv")
Univ
```

```python
#Standardise or normalise datasets
#normalisation of x. x is i in the below function. max min
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
df_norm = norm_func(Univ.iloc[:,1:])
df_norm
```

```python
#Normalising with SK learn minmax function easier than previous fn. no need to write formulas
#converting all our values bw 0 and 1 with min max function
from sklearn.preprocessing import MinMaxScaler
trans = MinMaxScaler()
data = pd.DataFrame(trans.fit_transform(Univ.iloc[:,1:]))
data
```

```python
#euclidean distance,linkage and dendrogram . method of linkage avarage and create dendrogram
from scipy.cluster.hierarchy import linkage
import scipy.cluster.hierarchy as sch # for creating dendrogram
#p = np.array(df_norm) # converting into numpy array format
z = linkage(df_norm, method="average",metric="euclidean")
plt.figure(figsize=(15, 5))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Index')
plt.ylabel('Distance')
sch.dendrogram(
    z,
    #leaf_rotation=6.,  # rotates the x axis labels
    #leaf_font_size=15.,  # font size for the x axis labels
)
plt.show()
```

```python
#Label every universities with a cluster a label (last col)
# agglomerative clustering and label them from 1 to 5 as indicated in dendrogram
#every universiti will get cluster label
from sklearn.cluster import AgglomerativeClustering
import warnings
warnings.filterwarnings('ignore')
h_complete = AgglomerativeClustering(n_clusters=5, linkage='average',affinity = "euclidean").fit(df_norm)

cluster_labels=pd.Series(h_complete.labels_)
cluster_labels
Univ['clust']=cluster_labels # creating a  new column and assigning it to new column
Univ
```

```python
# Group all universities based on cluster and calc mean
#from col1 as 0 col is names of uni
# based on clusters example clust 1 less SAT and exp
Univ.iloc[:,1:].groupby(Univ.clust).mean()
```
